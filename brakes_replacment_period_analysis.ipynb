```python
def make_review(parts_df):
    mx = list
    res = pd.DataFrame()
    for i in sep:
        frpd  = parts_df[parts_df['registry_vehicle_id']==i]
        milages = frpd['mileage'].unique()
        lst = []
        for v in milages:
            v = int(v)
            lst.append(v)
        lst.sort()
        if len(lst) == 0:
            pass
        else:
            counter = 1
            dic = {}
            for q in lst:
                dic[counter] = q
                counter+=1
                d = pd.DataFrame(dic, index=[0])
                d['vehicle_id'] = i
                res = res.append(d, ignore_index=True)
    res.columns = res.columns.astype(str)
    res = res.merge(cars, left_on='vehicle_id', right_on='registry_vehicle_id')
    res['vehicle'] = res['brand'] + '_' + res['model']


    one = res[['vehicle_id', 'vehicle', '1']]
    one = one.dropna().reset_index(drop=True)
    one = one.groupby(['vehicle'], as_index=False).agg({'1':'mean'})
    
    try:
        two = res[['vehicle_id', 'vehicle', '2']]
        two = two.dropna().reset_index(drop=True)
        two = two.groupby(['vehicle'], as_index=False).agg({'2':'mean'})
        one = one.merge(two, how= 'left', left_on='vehicle', right_on='vehicle')
    except:
        two = res[['vehicle_id', 'vehicle']]
        
    try:
        three = res[['vehicle_id', 'vehicle', '3']]
        three = three.dropna().reset_index(drop=True)
        three = three.groupby(['vehicle'], as_index=False).agg({'3':'mean'})
        one = one.merge(three, how= 'left', left_on='vehicle', right_on='vehicle')
    except:
        three = res[['vehicle_id', 'vehicle']]
    
    try:
        four = res[['vehicle_id', 'vehicle', '4']]
        four = four.dropna().reset_index(drop=True)
        four = four.groupby(['vehicle'], as_index=False).agg({'4':'mean'})
        one = one.merge(four, how= 'left', left_on='vehicle', right_on='vehicle')
    except:
        four = res[['vehicle_id', 'vehicle']]
    
    try:
        five = res[['vehicle_id', 'vehicle', '5']]
        five = five.dropna().reset_index(drop=True)
        five = five.groupby(['vehicle'], as_index=False).agg({'5':'mean'})
        one = one.merge(five, how= 'left', left_on='vehicle', right_on='vehicle')
    except:
        five = res[['vehicle_id', 'vehicle']]

    return one
```
